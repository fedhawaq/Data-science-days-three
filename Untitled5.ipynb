#**Step** 2: Data Cleaning

a) Filling missing ages with the median and  ensures statistical analysis

```python

df['Age'] = df['Age'].fillna(df['Age'].median())
```

b)Filling missing embarked rows by mode(most frequent value)

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

c) Cabin column is discarded because missing data is too high and it won’t give meaningful analysis without external data.

```python
df = df.drop(columns=['Cabin'], errors='ignore')
```

d) removing duplicate

```python
df = df.drop_duplicates()
```

#**Step** 3: Data Analysis

a) Survival rate by gender


```python

survival_by_gender = df.groupby('Sex')['Survived'].mean()
print("Survival rate by gender:\n", survival_by_gender)
```

b) Survival rate by class (Pclass)

```python

survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by class:\n", survival_by_class)
```

c)Average age per class

```python

average_age_class = df.groupby('Pclass')['Age'].mean()
print("\nAverage age per class:\n", average_age_class)
```

d) Survival rate by age group

```python

# Create age bins
bins = [0, 12, 18, 35, 60, 80]
labels = ['Child', 'Teen', 'Adult', 'Middle-aged', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Group by AgeGroup and calculate survival rate
survival_by_agegroup = df.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival rate by age group:\n", survival_by_agegroup)
```

#**Step** 4: Filtering

a) Female passengers who survived

```python

female_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print("Number of female survivors:", len(female_survived))
print(female_survived.head())  # Optional: preview first 5
```

b) Children who survived (Age ≤ 12)

```python

children_survived = df[(df['Age'] <= 12) & (df['Survived'] == 1)]
print("Number of children survivors:", len(children_survived))
print(children_survived.head())
```

c) Passengers in 1st class with high survival probability

```python

first_class_survived = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
print("Number of 1st class survivors:", len(first_class_survived))
```

#Step 5: **Insights**

1. Gender and Survival

When I look at the data, one thing stands out immediately: gender played the biggest role in survival. Women had a far better chance of surviving than men. In fact, about 74% of all women survived, while only around 19% of men made it.
This strongly reflects the evacuation rule followed on the Titanic: “women first.”

2. Impact of Passenger Class

 as we see in the analysis Passenger class also had a major influence on survival.
People traveling in 1st class had the highest survival rate, followed by 2nd class, while 3rd class passengers had the lowest chance of surviving.

A big reason for this is that wealthier passengers were located closer to the deck and the lifeboats, so they could reach them faster when the ship began sinking.

3. Were Children Prioritized?

Yes .
Children aged 12 and below had noticeably better survival chances compared to adults.
This matches historical accounts showing that children were given priority during lifeboat loading, especially when accompanied by women.

4. Group With the Highest Survival Probability

The group with the best chance of survival was:

➜ Female + 1st Class + Child

This combination benefited from all the key advantages:

Priority during evacuation (women)
Close access to lifeboats (1st class)
Special protection policies (children)

No other group had a survival rate as high.
